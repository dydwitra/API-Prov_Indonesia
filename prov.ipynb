```python
# ambil data provinsi
url_provinsi = "http://dev.farizdotid.com/api/daerahindonesia/provinsi"
response_provinsi = requests.get(url_provinsi)
data_provinsi =response_provinsi.json()

df_provinsi = pd.DataFrame(data_provinsi['provinsi'])
df_provinsi = df_provinsi[['id','nama']]
```

```python
#ambil data kabupaten /kota
df_kabkota = pd.DataFrame()
for index, row in df_provinsi.iterrows():
    id_provinsi = row['id']
    url_kabkota = f"https://dev.farizdotid.com/api/daerahindonesia/kota?id_provinsi={id_provinsi}"
    response_kabkota = requests.get(url_kabkota)
    data_kabkota = response_kabkota.json()

    df_temp = pd.DataFrame(data_kabkota['kota_kabupaten'])
    df_temp = df_temp[['id','nama','id_provinsi']]
    df_kabkota = pd.concat([df_kabkota, df_temp], ignore_index=True)
```

```python
#ambil data kecamatan /kota
df_kecamatan = pd.DataFrame()
for index, row in df_kabkota.iterrows():
    id_kabkota = row['id']
    url_kecamatan = f"https://dev.farizdotid.com/api/daerahindonesia/kecamatan?id_kota={id_kabkota}"
    response_kecamatan = requests.get(url_kecamatan)
    data_kecamatan = response_kecamatan.json()

    df_temp = pd.DataFrame(data_kecamatan['kecamatan'])
    df_temp = df_temp[['id','nama','id_kota']]
    df_kecamatan = pd.concat([df_kecamatan, df_temp], ignore_index=True)
```

```python
#ambil data kelurahan /desa
df_kelurahan = pd.DataFrame()
for index, row in df_kecamatan.iterrows():
    id_kecamatan = row['id']
    url_kelurahan = f"https://dev.farizdotid.com/api/daerahindonesia/kelurahan?id_kecamatan={id_kecamatan}"
    response_kelurahan = requests.get(url_kelurahan)
    data_kelurahan = response_kelurahan.json()

    df_temp = pd.DataFrame(data_kelurahan['kelurahan'])
    df_temp = df_temp[['id','nama','id_kecamatan']]
    df_kelurahan = pd.concat([df_kelurahan, df_temp], ignore_index=True)
```

```python
df = pd.concat([df_provinsi, df_kabkota, df_kecamatan, df_kelurahan], axis = 0, join ='inner', ignore_index=True)
df.to_excel("data_wilayah_indonesia.xlsx", index=False)
```
